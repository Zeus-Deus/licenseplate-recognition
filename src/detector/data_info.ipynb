```python
import os
import kagglehub
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
# Download latest version
path = kagglehub.dataset_download("fareselmenshawii/large-license-plate-dataset")

print("Path to dataset files:", path)

print("Gedownloade bestanden:")
for file in os.listdir(path):
    print(file)
```

```python
print("Inhoud van de 'labels' map:")
for file in os.listdir(os.path.join(path, 'labels')):
    print(file)

print("\nInhoud van de 'images' map:")
for file in os.listdir(os.path.join(path, 'images')):
    print(file)
```

```python
num_labels = len(os.listdir(os.path.join(path, 'labels')))
num_images = len(os.listdir(os.path.join(path, 'images')))

print(f"Aantal labelbestanden: {num_labels}")
print(f"Aantal afbeeldingen: {num_images}")
```

```python
# Get sub-directories first
image_dirs = os.listdir(os.path.join(path, 'images'))
# Select a random directory (train, test, or val)
random_dir = random.choice(image_dirs)
# Get image files from that directory
image_files = os.listdir(os.path.join(path, 'images', random_dir))
# Choose a random image file
if image_files:
    random_image = random.choice(image_files)
    
    img_path = os.path.join(path, 'images', random_dir, random_image)
    img = Image.open(img_path)
    
    plt.figure(figsize=(10, 8))
    plt.imshow(img)
    plt.title(f"Voorbeeld afbeelding: {random_dir}/{random_image}")
    plt.axis('off')
    plt.show()
```

```python
# Select 5 random images from the dataset and display their formats
random_images = []

for _ in range(5):
    # Choose random directory (train, test, or val)
    rand_dir = random.choice(image_dirs)
    # Get image files from that directory
    img_files = os.listdir(os.path.join(path, 'images', rand_dir))
    # Choose a random image file
    if img_files:
        rand_img = random.choice(img_files)
        img_path = os.path.join(path, 'images', rand_dir, rand_img)
        random_images.append((rand_dir, rand_img, img_path))

# Display info for each selected image
for idx, (dir_name, img_name, img_path) in enumerate(random_images, 1):
    img = Image.open(img_path)
    
    print(f"Image {idx}: {dir_name}/{img_name}")
    print(f"  Format: {img.format}")
    print(f"  Size: {img.size}")
    print(f"  Mode: {img.mode}")
    print(f"  Dimensions: {img.width}x{img.height}")
    print("-" * 50)
    
    plt.figure(figsize=(5, 4))
    plt.imshow(img)
    plt.title(f"Image {idx}: {dir_name}/{img_name}")
    plt.axis('off')
    plt.show()
```
